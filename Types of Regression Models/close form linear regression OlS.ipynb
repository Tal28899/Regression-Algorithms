```python
class Slr:
    # constructer
    def __init__(self):
        self.m=None
        self.b=None
    # Method to find m and b    
    def fit(self,X_train,y_train):
        nume=0
        deno=0
        for i in range(X_train.shape[0]):
            nume +=(y_train[i] - np.mean(y_train))*(X_train[i] - np.mean(X_train))
            deno += (X_train[i]-np.mean(X_train))**2
        self.m = nume/deno
        self.b = np.mean(y_train) - self.m *(np.mean(X_train))
        print(self.m)
        print(self.b)
     #Method that give a list of my model prediction   
    def predict(self,X_test):
        y_pred = []
        for i in range(X_test.shape[0]):
            prediction = self.m*(X_test[i]) + self.b
            y_pred.append(prediction)  
        return y_pred    
```
